```python
import pandas as pd
import numpy as np
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None)
```

```python
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", "drive-wheels", "engine-location", "wheel-base", " length", "width", "height", "curb-weight", "engine-type", "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]
df.columns = headers
```

```python
df.dropna(subset = ['price'], axis = 0, inplace = True )
```

```python
mean = df["compression-ratio"].mean()
df["compression-ratio"].replace(np.nan, mean)
```

```python
df["city-mpg"] = 235/df["city-mpg"]
df.rename(columns = {"city-mpg" : "city-L/100km"}, inplace = True)
```

```python
df["price"].replace('?', 0, inplace = True)
```

```python
df["price"] = df["price"].astype('float64')
df['price']
```

```python
bins = np.linspace(min(df['price']), max(df['price']), 4)
group = ['low', 'mid', 'high']
df['price-binned'] = pd.cut(df['price'], bins, labels = group, include_lowest = True)
```

```python
df2 = df['price-binned'].value_counts()
```

```python
df1 = pd.get_dummies(df['fuel-type'])
```

```python
df = pd.concat([df, df1], axis = 1)
df
```

```python
df_test = df[['drive-wheels', 'body-style', 'price']]
df_grp = df_test.groupby(['drive-wheels', 'body-style'], as_index = False).mean()
df_grp
```

```python
df_pivot = df_grp.pivot(index = 'drive-wheels', columns = 'body-style')
df_pivot
```

```python
import scipy.stats as stats
df_anova = df[['make', 'price']]
grp_anova = df_anova.groupby(['make'])
anova_res = stats.f_oneway(grp_anova.get_group('honda')['price'], grp_anova.get_group('subaru')['price'])
anova_res
```

```python
df['horsepower'] = df['horsepower'].replace('?', 0).astype('float64')
p_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
p_coef, p_value
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
x = df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
y = df[['price']]
lr.fit(x, y)
y_pred = lr.predict(x)
lr.score(x, y)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.regplot( x = 'highway-mpg', y = 'price', data = df)
```

```python
sns.residplot(df['highway-mpg'], df['price'])
```

```python
ax1 = sns.distplot(df['price'], hist = False, color = 'r', label = 'Actual Value')
sns.distplot(y_pred, hist = False, color = 'b', label = 'Fitted Value', ax = ax1)
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
Input = [('scale', StandardScaler()), ('poly', PolynomialFeatures(degree = 2)), ('model', LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(x, y)
y_pred2 = pipe.predict(x)
pipe.score(x, y)
```

```python
ax1 = sns.distplot(df['price'], hist = False, color = 'r', label = 'Actual Value')
sns.distplot(y_pred2, hist = False, color = 'b', label = 'Fitted Value', ax = ax1)
```

```python
scale2 = StandardScaler()
scale2.fit(x)
x_scale = scale2.transform(x)
pr = PolynomialFeatures(degree = 2)
pr.fit(x_scale)
x_pr = pr.transform(x_scale)
lm = LinearRegression()
lm.fit(x_pr, y)
y_pred3 = lm.predict(x_pr)
lm.score(x_pr, y)
```

```python
ax1 = sns.distplot(df['price'], hist = False, color = 'r', label = 'Actual Value')
sns.distplot(y_pred3, hist = False, color = 'b', label = 'Fitted Value', ax = ax1)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_pred)
```

```python
from sklearn.model_selection import train_test_split
xt, xte, yt, yte = train_test_split(x, y, test_size = 0.3, random_state = 0)
pipe.fit(xt, yt)
yp = pipe.predict(xte)
pipe.score(xt, yt)
```

```python
ax1 = sns.distplot(yte, hist = False, color = 'r', label = 'Actual Value')
sns.distplot(yp, hist = False, color = 'b', label = 'Fitted Value', ax = ax1)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(pipe, x, y, cv = 3)
np.mean(scores)
```

```python
from sklearn.model_selection import cross_val_predict
ym = cross_val_predict(pipe, x, y, cv = 3)
ym
```

```python
ax1 = sns.distplot(y, hist = False, color = 'r', label = 'Actual Value')
sns.distplot(ym, hist = False, color = 'b', label = 'Fitted Value', ax = ax1)
```

```python
rsqu = []
order = [1,2,3,4]
for n in order:
    p = PolynomialFeatures(degree = n)
    p.fit(xt)
    xt_p = p.transform(xt)
    p.fit(xte)
    xte_p = p.transform(xte)
    lm.fit(xt_p, yt)
    rsqu.append(lm.score(xte_p, yte))
    
rsqu
```
